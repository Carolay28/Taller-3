#### - Series de Taylor

Grafique las curvas de las series de Taylor de varios órdenes para los siguientes casos:

- $ln(x); x_0 = 1$

SOLUCIÓN:

$f'(x) = \frac{1}{x}$

$f''(x) = - \frac{1}{x^2}$

$f'''(x) = \frac{1}{x^3}$

$f(x) \approx f(x_0) + f´(x)(x - x_0) + \frac{f´´(x_0)}{2!} (x - x_0)^2 + \frac{f´´´(x_0)}{3!} (x - x_0)^3$

$f(x) \approx ln(1) + (x -1) - \frac{(x - 1)^2}{2} + \frac{2}{6}(x - 1)^3$

$f(x) \approx \frac{1}{3}(x - 1)^3 - \frac{(x - 1)^2}{2} + (x - 1)$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 5, 400)

f = np.log(x)                           
g = x - 1                            
h = -0.5*(x - 1)**2 + (x - 1)        

plt.figure(figsize=(10, 6))

plt.plot(x, f, color='green', label=r'$f(x) = \ln(x)$')
plt.plot(x, g, color='red', label=r'$g(x) = x - 1$')
plt.plot(x, h, color='blue', label=r'$h(x) = -\frac{(x-1)^2}{2} + x - 1$')

plt.grid(True)
plt.legend()
plt.title('Funciones y sus aproximaciones alrededor de $x=1$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axhline(0, color='black', linewidth=0.7)  
plt.axvline(0, color='black', linewidth=0.7)  

plt.xlim(-1, 5)
plt.ylim(-9, 10)

plt.show()
```

- $ln(x); x_0 = 10$

SOLUCIÓN:

$f'(x) = \frac{1}{x}$  

$f''(x) = -\frac{1}{x^2}$  

$f'''(x) = \frac{2}{x^3}$  

$f(x) \approx f(x_0) + f'(x_0)(x - x_0) + \frac{f''(x_0)}{2!}(x - x_0)^2 + \frac{f'''(x_0)}{3!}(x - x_0)^3$  

$f(x) \approx \ln(10) + \frac{1}{10}(x - 10) - \frac{1}{2(10)^2}(x - 10)^2 + \frac{2}{6(10)^3}(x - 10)^3$  

$f(x) \approx \ln(10) + 0.1(x - 10) - 0.005(x - 10)^2 + 0.000333(x - 10)^3$  

$f(x) \approx 2.3026 + 0.1(x - 10) - 0.005(x - 10)^2 + 0.000333(x - 10)^3$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 20, 400)

f = np.log(x)

g = np.log(10) + 0.1*(x - 10)

h = np.log(10) + 0.1*(x - 10) - 0.005*(x - 10)**2

k = np.log(10) + 0.1*(x - 10) - 0.005*(x - 10)**2 + 0.0003333*(x - 10)**3

plt.figure(figsize=(10, 6))

plt.plot(x, f, color='green', label=r'$f(x) = \ln(x)$')
plt.plot(x, g, color='red', label=r'$g(x) = \ln(10) + 0.1(x-10)$')
plt.plot(x, h, color='blue', label=r'$h(x) = \ln(10) + 0.1(x-10) - 0.005(x-10)^2$')
plt.plot(x, k, color='purple', label=r'$k(x) = \ln(10) + 0.1(x-10) - 0.005(x-10)^2 + 0.0003333(x-10)^3$')

plt.grid(True)
plt.legend()
plt.title('Funciones y sus aproximaciones alrededor de $x=10$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

plt.xlim(1, 20)
plt.ylim(-2, 6)

plt.show()
```

- $cos(x);  x_0 = 0$

$f'(x) = -\sin(x)$  

$f''(x) = -\cos(x)$  

$f'''(x) = \sin(x)$  

$f(x) \approx f(x_0) + f'(x_0)(x - x_0) + \frac{f''(x_0)}{2!}(x - x_0)^2 + \frac{f'''(x_0)}{3!}(x - x_0)^3$  

$f(x) \approx \cos(0) - \sin(0)(x - 0) - \frac{\cos(0)}{2!}(x - 0)^2 + \frac{\sin(0)}{3!}(x - 0)^3$  

$f(x) \approx 1 - 0(x) - \frac{1}{2}x^2 + 0$  

$f(x) \approx 1 - \frac{x^2}{2}$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 400)

f = np.cos(x)                  
g = 1                          
h = 1 - (x**2)/2               

plt.figure(figsize=(10, 6))

plt.plot(x, f, color='green', label=r'$f(x) = \cos(x)$')
plt.plot(x, [g]*len(x), color='red', label=r'$g(x) = 1$')
plt.plot(x, h, color='blue', label=r'$h(x) = 1 - \frac{x^2}{2}$')

plt.grid(True)
plt.legend()
plt.title('Función $\cos(x)$ y sus aproximaciones alrededor de $x=0$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

plt.xlim(-4, 4)
plt.ylim(-2, 2)

plt.show()
```

- $f(x) = \frac{1}{1-x} ; x_0 = 0.5$

$f'(x) = \frac{1}{(1 - x)^2}$  

$f''(x) = \frac{2}{(1 - x)^3}$  

$f'''(x) = \frac{6}{(1 - x)^4}$  

$f(x) \approx f(x_0) + f'(x_0)(x - x_0) + \frac{f''(x_0)}{2!}(x - x_0)^2 + \frac{f'''(x_0)}{3!}(x - x_0)^3$  

$f(x) \approx 2 + 4(x - 0.5) + 8(x - 0.5)^2 + 16(x - 0.5)^3$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 1, 400)

f = 1 / (1 - x)                                
g = 2 + 4*(x - 0.5)                            
h = 2 + 4*(x - 0.5) + 8*(x - 0.5)**2           
i = 2 + 4*(x - 0.5) + 8*(x - 0.5)**2 + 16*(x - 0.5)**3   

plt.figure(figsize=(10, 6))

plt.plot(x, f, color='green', label=r'$f(x) = \frac{1}{1 - x}$')
plt.plot(x, g, color='red', label=r'$g(x) = 2 + 4(x - 0.5)$')
plt.plot(x, h, color='blue', label=r'$h(x) = 2 + 4(x - 0.5) + 8(x - 0.5)^2$')
plt.plot(x, i, color='orange', label=r'$i(x) = 2 + 4(x - 0.5) + 8(x - 0.5)^2 + 16(x - 0.5)^3$')

plt.grid(True)
plt.legend()
plt.title('Aproximaciones de $\\frac{1}{1-x}$ alrededor de $x=0.5$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

plt.xlim(-2, 1)
plt.ylim(-10, 20)

plt.show()
```

#### - Polinomio de Lagrange

Encuentre el polinomio de Lagrange para los siguientes datos y grafique:

- $(0,0),(30,0.5),(60,\frac{\sqrt{3}}{2}),(90,1)$

$P(x) = y_0L_0(x) + y_1L_1(x) + y_2L_2(x) + y_3L_3(x)$

----------------------------------------------------------------------------------------------------------

$L_0(x) = \frac{(x - 30)(x - 60)(x - 90)}{(0 - 30)(0 - 60)(0 - 90)} = \frac{(x - 30)(x - 60)(x - 90)}{-162000}$

$L_1(x) = \frac{(x - 0)(x - 60)(x - 90)}{(30 - 0)(30 - 60)(30 - 90)} = \frac{x(x - 60)(x - 90)}{54000}$

$L_2(x) = \frac{(x - 0)(x - 30)(x - 90)}{(60 - 0)(60 - 30)(60 - 90)} = \frac{x(x - 30)(x - 90)}{-54000}$

$L_3(x) = \frac{(x - 0)(x - 30)(x - 60)}{(90 - 0)(90 - 30)(90 - 60)} = \frac{x(x - 30)(x - 60)}{162000}$

---------------------------------------------------------------------------------------------------------

$P(x) = 0 \cdot L_0(x) + 0.5 \cdot L_1(x) + \frac{\sqrt{3}}{2} \cdot L_2(x) + 1 \cdot L_3(x)$

$P(x) = \frac{0.5x(x - 60)(x - 90)}{54000} - \frac{\sqrt{3}\,x(x - 30)(x - 90)}{108000} + \frac{x(x - 30)(x - 60)}{162000}$

$P(x) = -6.054087146$ x $10^{-7} x^3 - 2.051287326$ x $10^{-4} x^2 + 0.02346539648 x$


```python
import numpy as np
import matplotlib.pyplot as plt

x_points = np.array([0, 30, 60, 90])
y_points = np.array([0, 0.5, np.sqrt(3)/2, 1])

coef = np.polyfit(x_points, y_points, 3)
f = np.poly1d(coef)

x = np.linspace(0, 90, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label=rf'$f(x)={coef[0]:.6e}x^3 + {coef[1]:.6e}x^2 + {coef[2]:.6e}x + {coef[3]:.6e}$')
plt.scatter(x_points, y_points, color='red', label='Puntos dados')
plt.grid(True)
plt.legend()
plt.title('Polinomio de Lagrange para los puntos : (0,0), (30,0.5), (60,√3/2), (90,1)')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.show()
```

- $(1,1),(2,2),(3,2)$

$P(x) = y_0L_0(x) + y_1L_1(x) + y_2L_2(x)$

----------------------------------------------------------------------------------------------------------

$L_0(x) = \frac{(x - 2)(x - 3)}{(1 - 2)(1 - 3)} = \frac{(x - 2)(x - 3)}{2}$

$L_1(x) = \frac{(x - 1)(x - 3)}{(2 - 1)(2 - 3)} = -(x - 1)(x - 3)$

$L_2(x) = \frac{(x - 1)(x - 2)}{(3 - 1)(3 - 2)} = \frac{(x - 1)(x - 2)}{2}$

----------------------------------------------------------------------------------------------------------

$P(x) = 1 \cdot L_0(x) + 2 \cdot L_1(x) + 2 \cdot L_2(x)$

$P(x) = \frac{(x - 2)(x - 3)}{2} - 2(x - 1)(x - 3) + \frac{2(x - 1)(x - 2)}{2}$

$P(x) = \frac{1}{2}(x^2 - 5x + 6) - 2(x^2 - 4x + 3) + (x^2 - 3x + 2)$

$P(x) = (-0.5)x^2 + (2.5)x - 1$

$P(x) = -\tfrac{1}{2}x^2 + \tfrac{5}{2}x - 1$


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -0.5 * x**2 + 2.5 * x - 1

x_points = np.array([1, 2, 3])
y_points = np.array([1, 2, 2])

x = np.linspace(0, 4, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', linewidth=2, label='f(x) = -0.5x² + 2.5x - 1')
plt.scatter(x_points, y_points, color='red', zorder=5, label='Puntos dados')

plt.grid(True)
plt.legend()
plt.title('Polinomio de Lagrange para los puntos (1,1), (2,2), (3,2)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.xlim(0, 4)
plt.ylim(0, 3)
plt.show()
```

- $(−2,5),(1,7),(3,11),(7,34)$

$P(x) = y_0L_0(x) + y_1L_1(x) + y_2L_2(x) + y_3L_3(x)$

----------------------------------------------------------------------------------------------------------

$L_0(x) = \dfrac{(x - 1)(x - 3)(x - 7)}{(-2 - 1)(-2 - 3)(-2 - 7)} = -\dfrac{(x - 1)(x - 3)(x - 7)}{135}$

$L_1(x) = \dfrac{(x + 2)(x - 3)(x - 7)}{(1 + 2)(1 - 3)(1 - 7)} = \dfrac{(x + 2)(x - 3)(x - 7)}{36}$

$L_2(x) = \dfrac{(x + 2)(x - 1)(x - 7)}{(3 + 2)(3 - 1)(3 - 7)} = -\dfrac{(x + 2)(x - 1)(x - 7)}{40}$

$L_3(x) = \dfrac{(x + 2)(x - 1)(x - 3)}{(7 + 2)(7 - 1)(7 - 3)} = \dfrac{(x + 2)(x - 1)(x - 3)}{216}$

----------------------------------------------------------------------------------------------------------

$P(x) = 5\cdot L_0(x) + 7\cdot L_1(x) + 11\cdot L_2(x) + 34\cdot L_3(x)$

$P(x) = 5\left(-\dfrac{(x - 1)(x - 3)(x - 7)}{135}\right) + 7\left(\dfrac{(x + 2)(x - 3)(x - 7)}{36}\right) + 11\left(-\dfrac{(x + 2)(x - 1)(x - 7)}{40}\right) + 34\left(\dfrac{(x + 2)(x - 1)(x - 3)}{216}\right)$

$P(x) = -\dfrac{5}{135}(x - 1)(x - 3)(x - 7) + \dfrac{7}{36}(x + 2)(x - 3)(x - 7) + \dfrac{11}{40}(x + 2)(x - 1)(x - 7) + \dfrac{34}{216}(x + 2)(x - 1)(x - 3)$

$P(x) = \dfrac{43}{1080}x^{3} + \dfrac{101}{540}x^{2} + \dfrac{793}{1080}x + \dfrac{1087}{180}$

$P(x) = 0.039814814x^{3} + 0.187037037x^{2} + 0.734259259x + 6.038888889$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 0.039814814 * x**3 + 0.187037037 * x**2 + 0.734259259 * x + 6.038888889

x_points = np.array([-2, 1, 3, 7])
y_points = np.array([5, 7, 11, 34])

x = np.linspace(-3, 8, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', linewidth=2,
         label='f(x) = 0.039814814x³ + 0.187037037x² + 0.734259259x + 6.038888889')
plt.scatter(x_points, y_points, color='red', zorder=5, label='Puntos dados')

plt.grid(True)
plt.legend()
plt.title('Polinomio de Lagrange para los puntos (-2,5), (1,7), (3,11), (7,34)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.xlim(-3, 8)
plt.ylim(0, 40)
plt.show()
```
